```python
import os
import cv2
import random
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from skimage.feature import hog
from sklearn.model_selection import train_test_split
```

```python
def extract_hog_features(image):
    hog_features = hog(
        image,
        orientations=9,
        pixels_per_cell=(8, 8),
        cells_per_block=(2, 2),
        visualize=False
    )
    return hog_features

def load_and_extract_features(directory, test_size=0.2, random_state=42):
    X, y = [], []
    for label in os.listdir(directory):
        label_dir = os.path.join(directory, label)
        for filename in os.listdir(label_dir):
            image_path = os.path.join(label_dir, filename)
            if filename == '.ipynb_checkpoints':
                continue
            img = cv2.imread(image_path)
            if img is None:
                continue
            img_resized = cv2.resize(img, (128, 128))
            img_gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)
            hog_features = extract_hog_features(img_gray)
            X.append(hog_features)
            y.append(label)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )
    return X_train, X_test, y_train, y_test

def train_random_forest(train_X, train_y):
    clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    clf.fit(train_X, train_y)
    return clf
```

```python
X_train, X_test, y_train, y_test = load_and_extract_features('Pokemon', test_size=0.3, random_state=42)
```

```python
rf_classifier = train_random_forest(X_train, y_train)
```

```python
predictions = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, predictions)

print("Classification Accuracy:", accuracy)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
