```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# RF Results
rf_results_df = pd.read_csv("rf_metrics.csv")
rf_results_df

# CNN Results
cnn_results_df = pd.read_csv("cnn_metrics.csv")
cnn_results_df

# NN Results
nn_results_df = pd.read_csv("nn_metrics.csv")
nn_results_df

# Combine all results
comparison_df = pd.concat(
    [rf_results_df, nn_results_df, cnn_results_df],
    ignore_index=True
)

comparison_df
```

```python
# Accuracy comparison
pivot_acc = comparison_df.pivot(index="Model", columns="Dataset", values="Accuracy")

pivot_acc.plot(kind="bar", figsize=(8,5))
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.ylim(0,1)
plt.xticks(rotation=0)
plt.legend(title="Dataset")
plt.tight_layout()
plt.show()
```
